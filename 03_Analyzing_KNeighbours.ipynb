```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import mglearn

# create dataset
X, y = mglearn.datasets.make_forge()

# create subplots (THIS WAS MISSING)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# loop
for n_neighbors, ax in zip([1, 3, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X, y)

    mglearn.plots.plot_2d_separator(
        clf, X, fill=True, eps=0.5, ax=ax, alpha=0.4
    )
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)

    ax.set_title(f"{n_neighbors} n_neighbors")
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")

axes[0].legend(loc=3)
plt.show()
```

```python

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer=load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(cancer.data,cancer.target,stratify=cancer.target)
training_accuracy=[]
testing_accuracy=[]
neighbour_setting=range(1,11)
for n_neighbors in neighbour_setting:
    clf=KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train,y_train)
    training_accuracy.append(clf.score(X_train,y_train))
    testing_accuracy.aappend(clf.score(X_test,y_test))
plt.plot(neighbour_setting,training_accuracy,label='training accuracy' )   
plt.plot(neighbour_setting,testing_accuracy,label='training accuracy' )                                               
```
